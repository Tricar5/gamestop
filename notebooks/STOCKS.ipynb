```python
import yfinance as yf
# Import the plotting library
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Get the data for the stock GME
data = yf.download('GME','2020-12-30','2021-03-01')
```

```python
data.head()
```

```python
sns.set()
plt.figure(figsize = (24,12))
sns.lineplot(x = "Date",
            y = "Adj Close",
            data =data, linewidth = 6, color='red')

#plt.title('GME Stock Prices ', fontsize = 48)

#plt.xlabel('Day of period', fontsize = 12)
#plt.ylabel('Price ($)', fontsize = 12)
plt.savefig('stock_prices.png')
plt.show()
```
